```python
# ============================================================
#  MÔ PHỎNG HUẤN LUYỆN ROBOT ĐIỀU HƯỚNG BẰNG DEEP Q-LEARNING
#  Robot học cách: đi tới đích, RẼ TRÁI / RẼ PHẢI, TRÁNH VẬT CẢN
#  và KHÔNG ĐÂM (đâm = "ngã" -> ván kết thúc).
#  Có hình ảnh trực quan: quay GIF robot ở nhiều mốc huấn luyện.
#  Chạy trực tiếp trên Google Colab (CPU là đủ).  Hỗ trợ AI.KTH
# ============================================================

# --- Cài thư viện (chạy 1 lần) ---
!pip install gymnasium imageio matplotlib -q

import os
os.environ["SDL_VIDEODRIVER"] = "dummy"   # cho phép vẽ hình không cần màn hình

import io, random, math
from collections import deque, namedtuple
import numpy as np
import torch, torch.nn as nn, torch.optim as optim
import matplotlib; matplotlib.use("Agg")
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, FancyArrow
import imageio.v2 as imageio
from IPython.display import Image, display

SEED = 7
random.seed(SEED); np.random.seed(SEED); torch.manual_seed(SEED)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Thiết bị:", device)

# ============================================================
#  PHẦN 1 — MÔI TRƯỜNG ROBOT ĐIỀU HƯỚNG 2D
# ============================================================
DIRS = {0:(-1,0), 1:(0,1), 2:(1,0), 3:(0,-1)}  # hướng: lên, phải, xuống, trái

def bfs_dist(grid, goal, N):
    """Khoảng cách đường-ngắn-nhất (né vật cản) từ mọi ô tới đích — dùng để định hình phần thưởng."""
    dist = np.full((N, N), np.inf); dist[goal] = 0; q = deque([goal])
    while q:
        r, c = q.popleft()
        for dr, dc in DIRS.values():
            nr, nc = r+dr, c+dc
            if 0<=nr<N and 0<=nc<N and not grid[nr,nc] and np.isinf(dist[nr,nc]):
                dist[nr,nc] = dist[r,c]+1; q.append((nr,nc))
    return dist

def make_map(N, ratio, seed):
    """Tạo bản đồ ngẫu nhiên CÓ ĐƯỜNG ĐI hợp lệ (kiểm tra bằng BFS)."""
    rng = np.random.default_rng(seed); start, goal = (0,0), (N-1,N-1)
    for _ in range(500):
        grid = (rng.random((N,N)) < ratio)
        grid[start] = grid[goal] = False
        dist = bfs_dist(grid, goal, N)
        if np.isfinite(dist[start]) and dist[start] >= N+2:
            return grid, dist, start, goal
    raise RuntimeError("Không tạo được bản đồ có đường đi")

class RobotNavEnv:
    """
    Trạng thái: vị trí (hàng, cột) + hướng quay.
    Hành động (rời rạc, hợp với DQN): 0 = đi thẳng, 1 = rẽ trái, 2 = rẽ phải.
    Đi thẳng đâm vào tường/vật cản  -> "ngã" (kết thúc, phạt nặng).
    Tới đích -> thưởng lớn.  Mỗi bước tốn một chút (đi đường ngắn).
    """
    def __init__(self, N=8, obstacle_ratio=0.20, seed=7, max_steps=80):
        self.N, self.max_steps = N, max_steps
        self.grid, self.dist_field, self.start, self.goal = make_map(N, obstacle_ratio, seed)
        self.n_actions = 3

    def _blocked(self, r, c):
        return not (0<=r<self.N and 0<=c<self.N) or self.grid[r, c]

    def _sensors(self):
        # cảm biến tương đối theo hướng robot: phía trước / bên trái / bên phải có bị chặn?
        out = []
        for d in (self.h % 4, (self.h-1) % 4, (self.h+1) % 4):
            dr, dc = DIRS[d]
            out.append(1.0 if self._blocked(self.r+dr, self.c+dc) else 0.0)
        return out

    def _obs(self):
        n = self.N - 1; gr, gc = self.goal
        return np.array([self.r/n, self.c/n, gr/n, gc/n,
                         np.sin(self.h*np.pi/2), np.cos(self.h*np.pi/2),
                         *self._sensors()], dtype=np.float32)

    def reset(self):
        self.r, self.c = self.start; self.h = 1; self.steps = 0
        self.trail = [(self.r, self.c)]
        return self._obs()

    def step(self, a):
        self.steps += 1
        prev_d = self.dist_field[self.r, self.c]; crashed = False
        if a == 0:                                   # đi thẳng
            dr, dc = DIRS[self.h]; nr, nc = self.r+dr, self.c+dc
            if self._blocked(nr, nc): crashed = True
            else: self.r, self.c = nr, nc; self.trail.append((nr, nc))
        elif a == 1: self.h = (self.h-1) % 4         # rẽ trái
        else:        self.h = (self.h+1) % 4         # rẽ phải

        reward, done, info = -0.05, False, {"status": "running"}
        if crashed:
            reward, done, info["status"] = -10.0, True, "crashed"
        elif (self.r, self.c) == self.goal:
            reward, done, info["status"] = 20.0, True, "goal"
        else:
            reward += (prev_d - self.dist_field[self.r, self.c])   # thưởng tiến gần đích
        if self.steps >= self.max_steps and not done:
            done, info["status"] = True, "timeout"
        return self._obs(), reward, done, info

# ============================================================
#  PHẦN 2 — MẠNG Q + EXPERIENCE REPLAY (giống DQN CartPole)
# ============================================================
env = RobotNavEnv(N=8, obstacle_ratio=0.20, seed=SEED, max_steps=80)
N_OBS, N_ACT = 9, env.n_actions

class QNetwork(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(nn.Linear(N_OBS,128), nn.ReLU(),
                                 nn.Linear(128,128), nn.ReLU(), nn.Linear(128,N_ACT))
    def forward(self, x): return self.net(x)

Transition = namedtuple("Transition", ("state","action","next_state","reward"))
policy_net = QNetwork().to(device)
target_net = QNetwork().to(device); target_net.load_state_dict(policy_net.state_dict())
optimizer = optim.AdamW(policy_net.parameters(), lr=1e-3, amsgrad=True)
memory = deque(maxlen=20000)
steps_done = 0
GAMMA, BATCH_SIZE, TAU = 0.99, 128, 0.005

def select_action(obs):
    global steps_done
    eps = 0.05 + 0.95*math.exp(-steps_done/3000); steps_done += 1
    if random.random() > eps:
        with torch.no_grad():
            return int(policy_net(torch.tensor(obs, device=device).unsqueeze(0)).argmax().item())
    return random.randrange(N_ACT)

def optimize_model():
    if len(memory) < BATCH_SIZE: return
    b = Transition(*zip(*random.sample(memory, BATCH_SIZE)))
    s  = torch.tensor(np.array(b.state), device=device)
    a  = torch.tensor(b.action, device=device).unsqueeze(1)
    r  = torch.tensor(b.reward, device=device, dtype=torch.float32)
    nfm = torch.tensor([x is not None for x in b.next_state], device=device, dtype=torch.bool)
    nfs = torch.tensor(np.array([x for x in b.next_state if x is not None]), device=device)
    sav = policy_net(s).gather(1, a)
    nsv = torch.zeros(BATCH_SIZE, device=device)
    with torch.no_grad(): nsv[nfm] = target_net(nfs).max(1).values
    loss = nn.SmoothL1Loss()(sav, (r + GAMMA*nsv).unsqueeze(1))
    optimizer.zero_grad(); loss.backward()
    torch.nn.utils.clip_grad_value_(policy_net.parameters(), 100); optimizer.step()

# ============================================================
#  PHẦN 3 — BỘ VẼ HÌNH (mỗi bước -> 1 ảnh) VÀ QUAY GIF
# ============================================================
def render_frame(status, step):
    # figsize + dpi cố định -> mọi khung hình CÙNG kích thước (bắt buộc để ghép GIF)
    fig, ax = plt.subplots(figsize=(4, 4.4), dpi=60); Nn = env.N
    fig.subplots_adjust(left=0.04, right=0.96, top=0.90, bottom=0.04)
    ax.set_xlim(0,Nn); ax.set_ylim(0,Nn); ax.set_aspect("equal"); ax.invert_yaxis()
    ax.set_xticks(range(Nn+1)); ax.set_yticks(range(Nn+1))
    ax.grid(True, color="#ddd", lw=0.5); ax.set_xticklabels([]); ax.set_yticklabels([])
    for r in range(Nn):                                  # vật cản
        for c in range(Nn):
            if env.grid[r,c]: ax.add_patch(Rectangle((c,r),1,1,color="#333"))
    gr,gc = env.goal                                     # đích
    ax.add_patch(Rectangle((gc,gr),1,1,color="#2ca02c"))
    ax.text(gc+0.5,gr+0.5,"ĐÍCH",ha="center",va="center",color="white",fontsize=7,weight="bold")
    if len(env.trail) > 1:                               # vệt đường đã đi
        ys=[p[0]+0.5 for p in env.trail]; xs=[p[1]+0.5 for p in env.trail]
        ax.plot(xs,ys,"-",color="#1f77b4",alpha=0.4,lw=2)
    dr,dc = DIRS[env.h]; bx,by = env.c+0.5, env.r+0.5    # robot
    col = "#d62728" if status=="crashed" else "#1f77b4"
    ax.add_patch(plt.Circle((bx,by),0.3,color=col,zorder=5))
    ax.add_patch(FancyArrow(bx,by,dc*0.45,dr*0.45,width=0.08,color="black",
                            zorder=6,length_includes_head=True,head_width=0.25))
    tt = {"running":"Đang đi","crashed":"ĐÂM! (ngã)","goal":"TỚI ĐÍCH!","timeout":"Hết giờ"}[status]
    ax.set_title(f"Bước {step} | {tt}", fontsize=10)
    buf = io.BytesIO(); fig.savefig(buf, format="png"); plt.close(fig)
    buf.seek(0); return imageio.imread(buf)

def record_gif(filename, trained=True, eval_eps=0.05):
    """Quay một ván rồi lưu GIF. trained=False: đi ngẫu nhiên (lúc chưa học)."""
    o = env.reset(); frames = [render_frame("running", 0)]; status = "timeout"; last = 0
    for k in range(1, env.max_steps+1):
        if (not trained) or random.random() < eval_eps:
            a = random.randrange(N_ACT)
        else:
            with torch.no_grad():
                a = int(policy_net(torch.tensor(o, device=device).unsqueeze(0)).argmax().item())
        o, r, d, info = env.step(a); last = k
        frames.append(render_frame(info["status"] if d else "running", k))
        if d: status = info["status"]; break
    imageio.mimsave(filename, frames, fps=4)
    return status, last

# ============================================================
#  PHẦN 4 — HUẤN LUYỆN + QUAY VIDEO TẠI CÁC MỐC
# ============================================================
NUM_EPISODES = 500
RECORD_AT = [0, 150, 350, NUM_EPISODES]   # các mốc episode sẽ quay video
recordings = {}; episode_rewards = []; goals = 0

# Quay TRƯỚC khi học (đi ngẫu nhiên)
st, ln = record_gif("robot_ep000.gif", trained=False)
recordings[0] = ("robot_ep000.gif", st, ln)
print(f"[Quay] Trước khi học (ngẫu nhiên): {st} sau {ln} bước")

for ep in range(1, NUM_EPISODES+1):
    o = env.reset(); total = 0
    while True:
        a = select_action(o)
        o2, r, d, info = env.step(a); total += r
        next_state = None if (d and info["status"] != "timeout") else o2
        memory.append(Transition(o, a, next_state, r)); o = o2
        optimize_model()
        tsd = target_net.state_dict(); psd = policy_net.state_dict()
        for k in psd: tsd[k] = psd[k]*TAU + tsd[k]*(1-TAU)
        target_net.load_state_dict(tsd)
        if d:
            episode_rewards.append(total)
            if info["status"] == "goal": goals += 1
            break
    if ep % 50 == 0:
        print(f"Episode {ep:3d} | reward TB 50ep {np.mean(episode_rewards[-50:]):6.2f} | tổng lần tới đích {goals}")
    if ep in RECORD_AT:
        fn = f"robot_ep{ep:03d}.gif"; st, ln = record_gif(fn, trained=True)
        recordings[ep] = (fn, st, ln)
        print(f"[Quay] Sau {ep} episode: {st} sau {ln} bước")

print(f"\nHuấn luyện xong! Tổng số lần tới đích: {goals}/{NUM_EPISODES}")

# --- Đường cong học ---
plt.figure(figsize=(10,4))
plt.plot(episode_rewards, alpha=0.35, label="Reward mỗi episode")
w = 30
if len(episode_rewards) >= w:
    ma = np.convolve(episode_rewards, np.ones(w)/w, mode="valid")
    plt.plot(range(w-1, len(episode_rewards)), ma, "r", lw=2, label=f"TB trượt {w} ep")
plt.axhline(20, color="green", ls="--", alpha=0.5, label="Mức thưởng khi tới đích")
plt.xlabel("Episode"); plt.ylabel("Tổng reward"); plt.legend(); plt.grid(True, alpha=0.3)
plt.title("Robot học điều hướng — đường cong reward"); plt.show()

# --- Hiển thị video qua các mốc ---
print("\n===== ROBOT QUA CÁC MỐC HUẤN LUYỆN =====")
TT = {"crashed":"đâm vào vật cản","goal":"tới đích","timeout":"hết giờ chưa tới"}
for ep in sorted(recordings):
    fn, st, ln = recordings[ep]
    nhan = "TRƯỚC khi học (ngẫu nhiên)" if ep == 0 else f"Sau {ep} episode"
    print(f"\n▶ {nhan} — kết quả: {TT[st]} (sau {ln} bước):")
    display(Image(filename=fn))
```
